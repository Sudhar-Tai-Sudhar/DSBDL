```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set_style("whitegrid")
```

```python
# Create histograms for each numeric feature
plt.figure(figsize=(10, 6))
df.iloc[:, :-1].hist(figsize=(10, 6), bins=15, edgecolor='black', color='skyblue')
plt.suptitle("Histograms of Numeric Features", fontsize=14)
plt.show()
```

```python
# Create box plots for each numeric feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.iloc[:, :-1], palette="Set2")
plt.title("Box Plots of Numeric Features", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# Create box plots for each feature grouped by species
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df["Species"], y=df[column], palette="Set3")
    plt.title(f"Distribution of {column} by Species")
```

```python
# Create box plots for each feature grouped by species without using a loop
plt.figure(figsize=(6, 6))
sns.boxplot(x=df["Species"], y=df["SepalLengthCm"], palette="Set3")
plt.title("Distribution of SepalLengthCm by Species")
plt.show()

plt.figure(figsize=(6, 6))
sns.boxplot(x=df["Species"], y=df["SepalWidthCm"], palette="Set3")
plt.title("Distribution of SepalWidthCm by Species")
plt.show()

plt.figure(figsize=(6, 6))
sns.boxplot(x=df["Species"], y=df["PetalLengthCm"], palette="Set3")
plt.title("Distribution of PetalLengthCm by Species")
plt.show()

plt.figure(figsize=(6, 6))
sns.boxplot(x=df["Species"], y=df["PetalWidthCm"], palette="Set3")
plt.title("Distribution of PetalWidthCm by Species")
plt.show()
```
